```python
import pandas as pd
```

```python
files = !ls -t figures/ | grep '_epi'|grep -v animate
```

```python
df = pd.DataFrame(files, columns=['files'])
```

```python
df['episode'] = df.files.apply(lambda x : int(x.split('_')[0]) )
```

```python
df['turn'] = df.files.apply(lambda x : int(x.split('_')[2]) )
```

```python
df['flash'] = df.files.apply(lambda x : int(x.split('_')[5]) )
```

```python
df_grouped = df.groupby('episode')
```

```python
for k, v in df_grouped:
    print (k)
    cmd1 = "rm -r figures/{k} ;mkdir figures/{k}".format(k=k)
    !$cmd1
    v_modified =  v.sort_values(['turn', 'flash']).copy()['files'].values.tolist()

    cmd_create_gif = "cd figures/; convert -delay 15 -loop 0 {files}  {k}_episode__animated.gif".format(k=k,files=' '.join(v_modified))
    #print (cmd_create_gif)
    !$cmd_create_gif
    
    
    
```
